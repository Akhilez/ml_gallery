```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
image_path = '/Users/akhil/code/ml_gallery/ml_py/data/pupil/test/images/C1_S1_I1.tiff'
```

```python
img = cv2.imread(image_path)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
_, thresh = cv2.threshold(img, 15, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours_img = cv2.drawContours(thresh, contours, -1, (0,255,0), 3)
```

```python
plt.imshow(contours_img)
```

```python
circles = cv2.HoughCircles(contours_img, cv2.HOUGH_GRADIENT, 2, contours_img.shape[0] / 2)
```
