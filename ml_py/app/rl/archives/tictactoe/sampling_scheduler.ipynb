```python
import torch
from torch.nn import functional as F
import numpy as np
import matplotlib.pyplot as plt
```

```python
seq = np.arange(1, 10001)
```

```python
seq_ln = np.log(seq)
```

```python
seq_ln_inv = (1 / seq_ln) * 5
seq_ln_inv[-1]
```

```python
plt.plot(seq_ln_inv)
plt.show()
```

```python
def plot(x, show=True):
    plt.plot(x)
    plt.ylim([0, 1])
    if show:
        plt.show()
```

```python
x = torch.tensor([4, 7, 3,2, 6,8, 9,4,2 ,6]).float()
xs = F.softmax(x, 0)
plot(xs)
```

```python
xt = F.gumbel_softmax(x, tau=0.7, dim=0)
plot(xt)
```

```python
for i in range(1, 10):
    plot(F.gumbel_softmax(x, tau=2, dim=0), show=False)
plt.show()
```
